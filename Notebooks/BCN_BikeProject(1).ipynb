# Prepare DataSets

```
import pandas as pd
import numpy as np
import geopandas as gp
import seaborn as sns
import rtree
```

```
city_edges='/content/drive/My Drive/Colab Notebooks/Project/city_edges_WGS84.csv'
```

```
city_edges=pd.read_csv(city_edges).iloc[:,1:]
city_edges.head()
```

```
type(city_edges)
```

```
from shapely import wkt

city_edges.geometry=city_edges.geometry.apply(wkt.loads)
```

```
city_edges_geo=gp.GeoDataFrame(city_edges,geometry='geometry',crs='WGS 84')
city_edges_geo.plot()
```

```
city_edges_geo.crs
```

```
#EDGES WITH 3043 CRS (METERS MEASURE)
edges=city_edges_geo.to_crs('EPSG:3043')
edges.head()
```

```
barris_geojson='/content/drive/My Drive/Colab Notebooks/Project/barris.geojson'
```

```
barris=gp.read_file(barris_geojson)
barris.head()
```

```
barris.plot()
```

```
barris.crs
```

```
barris=barris.to_crs("EPSG:3043")
barris.crs
```

```
barris=barris.loc[barris.geometry.type=='Polygon']
```

```
#prepare only barri name and geometry
barris=barris[['NOM','geometry']]
```

SPATIAL JOIN

```
edges_barri=gp.sjoin(left_df=edges,right_df=barris,how='left')
edges_barri.shape
```

```
edges_barri.head()
```

question: How can we meassure now the new distance of each street

```
edges_barri.loc[edges_barri.NOM=='Sant Antoni'].plot()
```

```
#Add buffer to geometry
#10 mts
edges_barri['geo_buf']=edges_barri.geometry.buffer(10)
edges_barri.set_geometry('geo_buf').plot()
```

```
edges_barri['geometry_original']=edges_barri.geometry
edges_barri['geometry']=edges_barri.geo_buf
edges_barri_buf=edges_barri.set_geometry('geometry')
edges_barri_buf.loc[edges_barri_buf.NOM=='Sant Antoni'].plot()
```

IMPORT ACCIDENTS

```
bike_acc='/content/drive/My Drive/Colab Notebooks/Project/bike_acc_16-18.csv'
```

```
accidents=pd.read_csv(bike_acc).iloc[:,1:]
accidents.head()
```

```
accidents_geo=gp.GeoDataFrame(accidents,geometry=gp.points_from_xy(accidents.Longitud,accidents.Latitud))
accidents_geo.crs={'init' :'epsg:4326'}
```

```
accidents_geo=accidents_geo.to_crs("EPSG:3043")
```

```
accidents_geo.plot()
```

```
accidents_geo.shape
```

```
accidents_geo=accidents_geo[['barri_carrer','geometry']]
accidents_geo.head()
```

Count accidents per street

```
edges_barri_buf.drop('index_right',axis=1,inplace=True)
```

```
edges_barri_buf.head()
```

```
#reset index an get another one to identify every street-barri
edges_barri_buf.drop('index',axis=1,inplace=True)
edges_barri_buf.reset_index(inplace=True)
```

```
edges_barri_buf.head()
```

```
#edges_barri_buf.drop('level_0',axis=1,inplace=True)
```

```

edges_with_accidents=gp.sjoin(left_df=edges_barri_buf, right_df=accidents_geo, how='left')
```

```
edges_with_accidents.head()
```

```
edges_with_accidents.shape,edges_barri_buf.shape
```

```
edges_with_accidents.loc[~edges_with_accidents.index_right.isna()]
```

```
edges_w_acc=edges_with_accidents.loc[~edges_with_accidents.index_right.isna()]
acc_per_street=edges_w_acc.index.value_counts().reset_index()
acc_per_street.columns=['id','acc_nbr']
acc_per_street.head()
```

```
edges_barri_acc=edges_barri_buf.merge(acc_per_street,left_on='index',right_on='id',how='left')
edges_barri_acc.acc_nbr.fillna(0,inplace=True)
edges_barri_acc.shape
```

```
import matplotlib.pyplot as plt

edges_barri_acc.plot(column='acc_nbr',legend=True,figsize=(15,8))
```

```
edges_barri_acc.loc[edges_barri_acc.acc_nbr>0].plot(column='acc_nbr',legend=True,figsize=(15,8))
```

```
edges_barri_acc.loc[edges_barri_acc.acc_nbr>1].plot(column='acc_nbr',legend=True,figsize=(15,8))
```

```
edges_barri_acc.acc_nbr.value_counts(normalize=True)*100
```

```
edges_barri_acc.acc_nbr.value_counts()
```

Count trees per street

```
trees='/content/drive/My Drive/Colab Notebooks/Project/trees.csv'
```

```
trees_data=pd.read_csv(trees)
trees_data.head()
```

```
trees_geo=gp.GeoDataFrame(trees_data,geometry=gp.points_from_xy(trees_data.LONGITUD_WGS84,trees_data.LATITUD_WGS84))
trees_geo.crs={'init' :'epsg:4326'}
```

```
trees_geo=trees_geo.to_crs("EPSG:3043")
```

```
trees_geo.plot(column='ALCADA',figsize=(20,15),legend=True)
```

```
trees_geo.ALCADA.value_counts(normalize=True)
```

```
trees_geo.loc[trees_geo.ALCADA=='GRAN'].plot(figsize=(20,15))
```

```
trees_geo.loc[trees_geo.ALCADA=='EXEMPLAR'].plot(figsize=(20,15))
```

```
trees_geo_f=trees_geo.loc[trees_geo.ALCADA.isin(['GRAN','EXEMPLAR'])]
trees_geo_f=trees_geo_f[['ADRECA','geometry']]
```

```
edges_trees=gp.sjoin(left_df=edges_barri_acc, right_df=trees_geo_f,how='left')
edges_trees.head()
```

```
edges_trees=edges_trees.loc[~edges_trees.ADRECA.isna()]
trees_per_street=edges_trees.index.value_counts().reset_index()
trees_per_street.columns=['id','trees_cnt']
trees_per_street.head()
```

```
edges_barri_acc_trees=edges_barri_acc.merge(trees_per_street,left_on='index',right_on='id',how='left')
edges_barri_acc_trees.trees_cnt.fillna(0,inplace=True)
edges_barri_acc_trees.shape
```

```

edges_barri_acc_trees.plot(column='trees_cnt',legend=True,figsize=(15,8))
```

```
edges_barri_acc_trees.loc[edges_barri_acc_trees.trees_cnt>0].plot(column='trees_cnt',legend=True,figsize=(15,8))
```

```
edges_barri_acc_trees.trees_cnt.plot(kind='hist')
```

```
edges_barri_acc_trees.trees_cnt.value_counts(normalize=True).cumsum()
```

```
edges_barri_acc_trees.trees_cnt.sum()
```

```
edges_barri_acc_trees.loc[edges_barri_acc_trees.trees_cnt>20].plot(column='trees_cnt',legend=True,figsize=(15,8))
```

# Add distance as measure

```
path='/content/drive/My Drive/Colab Notebooks/Project/'

edges_barri_acc_trees=pd.read_csv(path+'edges_trees_acc.csv')
```

```
edges_barri_acc_trees=gp.GeoDataFrame(edges_barri_acc_trees)
edges_barri_acc_trees.crs={'init' :'epsg:3043'}

edges_barri_acc_trees.crs
```

```
from shapely.wkt import loads

edges_barri_acc_trees.geometry_original=edges_barri_acc_trees.geometry_original.apply(lambda x: loads(x))
edges_barri_acc_trees.set_geometry('geometry_original').plot()
```

```
edges_barri_acc_trees.geometry_original[0].length
```

```
edges_barri_acc_trees['length_cutted']=edges_barri_acc_trees.geometry_original.apply(lambda x: x.length)
```

```
edges_barri_acc_trees.to_csv(path+'edges_trees_acc.csv')
```
